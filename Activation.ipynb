```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
w0 = -0.5
x = [1, 1]
w = [0.5, 0.5]

w = np.array(w)
x = np.array(x)

s = w0 + np.dot(w, x)
print(s)
```

```python
w0 = -0.5
x = [0, 0]
w = [0.5, 0.5]

w = np.array(w)
x = np.array(x)

s = w0 + np.dot(w, x)
print(s)
```

```python
def and_gate(x1, x2):
    w0 = -0.5 # bias
    w1 = 0.4
    w2 = 0.4
    
    th = 0.0
    
    s = w0+(w1*x1)+(w2*x2)
    if s >= th:
        result = 1
    else:
        result = 0
    print("output y={} [s={:.2f}]".format(result, s))
    
and_gate(0, 0)
and_gate(0, 1)
and_gate(1, 0)
and_gate(1, 1)
```

```python
def binary_step(x):
    result = 1 if x >= 0 else 0
    return result

input = np.arange(-7, 7, 0.15)
output = [binary_step(x) for x in input]
plt.plot(input ,output)
plt.xlabel("x (input)")
plt.ylabel("y (output)")
```

```python
def linear05(x):
    return 0.5 * x

def linear1(x):
    return 1 * x

def linear2(x):
    return 2 * x

input = np.arange(-7, 7, 0.15)
output05 = [linear05(x) for x in input]
output1 = [linear1(x) for x in input]
output2 = [linear2(x) for x in input]

plt.plot(output05)
plt.plot(output1)
plt.plot(output2)
plt.xlabel("x (input)")
plt.ylabel("y (output)")
```

```python
def sigmoid(x):
    return 1.0/(1.0+math.exp(-x))

input = np.arange(-7, 7, 0.15)
output = [sigmoid(x) for x in input]

plt.plot(output)
plt.xlabel("x (input)")
plt.ylabel("y (output)")
```

```python
def ReLu(x):
    return max(0.0, x)

input = np.arange(-7, 7, 0.15)
output = [ReLu(x) for x in input]

plt.plot(output)
plt.xlabel("x (input)")
plt.ylabel("y (output)")
```

```python
def leaky_ReLu(x):
    if x < 0:
        return 0.1 * x
    else:
        return x

input = np.arange(-7, 7, 0.5)
output = [leaky_ReLu(x) for x in input]

plt.plot(output)
plt.xlabel("x (input)")
plt.ylabel("y (output)")

plt.axhline(y=0, color='r', linestyle='--', lw=1.2)
# plt.plot(input, output, c='b', lw=2)
```
