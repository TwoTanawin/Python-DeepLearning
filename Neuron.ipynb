<h3>Applied AND Gate</h3>

```python
import numpy as np
```

```python
x = [[0,0],[0,1],[1,0],[1,1]]
x_train = np.array(x)

y = [0,0,0,1]
y_train = np.array(y)
```

```python
from sklearn.linear_model import Perceptron

# model = Perceptron()
model = Perceptron(random_state=0 ,verbose=True)

model.fit(x_train, y_train)
```

```python
print('score : ',model.score(x_train, y_train))
print('predict : ', model.predict(x_train))
print('target : ',np.array(y))
```

```python
print('weight [w1, w2] : ',model.coef_[0])
print('w0 (Bias) : ', model.intercept_[0])
```

<h3>XOR Multilayer</h3>

```python
x = [[0,0],[0,1],[1,0],[1,1]]
x_train = np.array(x)

y = [0, 1, 1, 0]
y_train = np.array(y)
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(4), random_state=1, verbose=True ,max_iter=1000)

model.fit(x_train, y_train)
```

```python
print('score : ',model.score(x_train, y_train))
print('predict : ', model.predict(x_train))
print('target : ',np.array(y))
```

```python
[coef.shape for coef in model.coefs_]
```

```python
print(model.coefs_) #weight
```

```python
print(model.intercepts_) # Bias
```

```python
import matplotlib.pyplot as plt
import mglearn

plt.figure(figsize=(5,4))
mk = {0:'o', 1:'s'}

mglearn.plots.plot_2d_separator(model, x_train, fill=True, eps=1.5, alpha=.28)
mglearn.discrete_scatter(x_train[:, 0], x_train[:, 1], y, s=12, markers=mk)

plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper left')
plt.title('MLP Perceptron')
```
